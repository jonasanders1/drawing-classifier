```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import os

# List of selected classes
classes = ['circle', 'square', 'triangle', 'cat', 'dog', 'bird', 'airplane', 'car', 'house', 'star', 'umbrella']

# Initialize a dictionary to store the number of samples in each class
class_counts = {}

# Initialize a dictionary to store sample images from each class
class_samples = {}
```

```python
# Load data and count samples
for class_name in classes:
    # Load the .npy file
    class_data = np.load(f'./npy_files/{class_name}.npy')
    
    # Store the number of samples
    class_counts[class_name] = len(class_data)
    
    # Store 3 random samples from the class
    random_indices = np.random.choice(len(class_data), 3, replace=False)
    class_samples[class_name] = class_data[random_indices]

# Print the number of samples in each class
print("Number of samples in each class:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}")
```

```python
# Plot 3 samples from each class
for class_name, samples in class_samples.items():
    print(f"\nClass: {class_name}")
    plt.figure(figsize=(10, 3))
    for i, sample in enumerate(samples):
        # Reshape the sample to 28x28 (if flattened)
        sample_image = sample.reshape(28, 28)
        
        # Plot the image
        plt.subplot(1, 3, i + 1)
        plt.imshow(sample_image, cmap='gray')
        plt.title(f"Sample {i + 1}")
        plt.axis('off')
    plt.show()
```
